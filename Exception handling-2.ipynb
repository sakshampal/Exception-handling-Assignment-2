```python
Q1. Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.

Ans: 1. The Exception class is like a blueprint or template for creating different types of errors in Python. 
        When we create a custom exception, we use the Exception class as the starting point for our own exception class.
    
     2. Exception class is like a parent, and our custom exception is like its child. The parent class provides all the important features 
        and behaviors that make handling errors in Python easier. By inheriting from the Exception class, our custom exception can use and
        benefit from these features.
        
     3. Using the Exception class as the base for our custom exception ensures that our custom exception works well with the existing error
        handling system in Python. It makes our custom exception compatible with try-except blocks and allows us to handle it in a consistent way.
        
     4. the Exception class provides helpful methods and attributes that allow us to customize our error messages and handle errors more effectively.
        We can override these methods to provide meaningful information about the error and use the attributes to get more insights into the error
        context.
```

```python
Q2. Write a python program to print Python Exception Hierarchy.
```

```python
def print_exception_hierarchy(exception_class , indent = 0):
    print(' ' * indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass , indent + 4)
        
print_exception_hierarchy(Exception)        
    
```

```python
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans : The LookupError class in Python is used as the base class for exceptions that occur when a lookup or indexing operation fails.
      It provides a common superclass for exceptions related to searching, indexing, or retrieving elements from a collection or sequence.
      Two examples of errors defined in the LookupError class are KeyError and IndexError.
```

```python
#KeyError

try:
    d = {"name": "saksham" , "age":23 , "course": "data science"}
    print(d["graduation"])
    
except KeyError as e:
    print("Error: The specified error can not exist",e)
```

```python
#Index Error

try:
    l = [1,2,3,"saksham",12.2,True]
    print(l[6])
except IndexError as e:
    print(e)
```

```python
Q5. Explain ImportError?

Ans: ImportError is an exception class in Python that is raised when an import statement fails to locate or load a module.
     It is a generic exception used for various import-related errors.
    
    When an import statement is executed, Python searches for the specified module in a set of directories defined by the system's
    module search path. If the module is not found or cannot be imported correctly, an ImportError is raised.
```

```python
#Ex: 

try :
    import saksham
except ImportError as e:
    print(e)
```

```python
 What is ModuleNotFoundError?
    
Ans: ModuleNotFoundError is a subclass of ImportError that specifically indicates that the requested module could not be found
```

```python
# Ex:
try :
    import saksham_pal
    
except ModuleNotFoundError as e :
    print(e)
```
